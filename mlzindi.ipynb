```python
from scipy.stats import chi2_contingency
```

```python
def cramers_v(x,y):
    #create a contigency table
    contingency_table=pd.crosstab(x,y)
    chi2_statistic, p_value, dof, expected = chi2_contingency(contingency_table)
    #calculate cramers v
    n=contingency_table.sum().sum()
    phi2=chi2_statistic/n
    r,k=contingency_table.shape
    phi2corr=max(0,phi2-((k-1)*(r-1))/(n-1))
    rcorr=r-(r-1)*(r-2)/(n-1)
    kcorr=k-(k-1)*(k-2)/(k-1)
    v=np.sqrt(phi2corr/min((kcorr-1,rcorr-1)))
    return v
    #categorical features
categorical_columns=['country','location_type','cellphone_access','gender_of_respondent','education_level']
target='bank_account'
cramers_value=[]

for category in categorical_columns:
    v_value=cramers_v(cleaned_data[category],cleaned_data[target])
    #print(f"{category} vs {target}: Cramer's V = {v_value:.4f}")
    cleanv_value = round(v_value, 4)
    cramers_value.append(float(cleanv_value))
cramers_value
```

MAJOR INSIGHTS
1.Year-Financial inclusion programs may evolve over time hence
       more people might have bank accounts because of government efforts, mobile money expansion

2.Age of respondents-Middle-aged adults (maybe 25-50) might be more likely to have bank accounts for work, loans, saving.
                     Very young or very old people might have fewer accounts.
                     
3.Educational level-Higher education often means better job opportunities and higher financial literacy, leading 
                     to more bank account ownership.

4.Country-Banking infrastructure varies by country.
          For example, Kenya has a strong mobile banking system (M-Pesa). Rwanda has been pushing financial inclusion. Tanzania and Uganda have different adoption rates thus affecting the usage or ownership of a bank account.

5.Cellphone access-In East Africa, mobile banking is a big driver of financial inclusion.
                   People with phones are more likely to use mobile money services or mobile banking apps linked to formal banks.

6.Gender of respodent-Gender gaps in financial inclusion exist.
                     In some areas, men may have more bank accounts due to cultural, economic, or legal reasons.

7.Location type-Urban dwellers usually have more access to banks, ATMs, mobile networks.
               Rural people might rely more on cash and may have fewer banks nearby.
